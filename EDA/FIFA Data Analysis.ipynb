```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('fifa.csv')
df.head()
```

```python
df.shape
# This dataset has 18207 rows and 60 columns.
```

```python
df.info()
# Data has only object,float and integer values.
```

```python
df.describe()
# Here as you can notice mean values of most columns are greater than median value of each column which is represented by 50% in index column.
# There are total 3932 players in this dataset
```

```python
# Histogram

plt.hist(df['Age'])
plt.xlabel('Players Age')
plt.ylabel('Number of players')
plt.grid(True)
plt.show()

# Players age is mostly in range from 20 to 35 
# The average age of players is 25 
# Maximum No of players are aged 24
# The youngest players age is 16 and the oldest players age is 47
```

```python
# Histogram

plt.hist(df['Overall'])
plt.xlabel('Players Rating')
plt.ylabel('Number of players')
plt.grid(True)
plt.show()

best = df.loc[df['Overall'].idxmax()]
print("The best player in FIFA 19 is", df['Overall'].max(), "overall. His name is", best['Name'], 
      'he is from',best['Nationality'],'and plays for',best['Club'],'.')
print('The median rating of a player on FIFA 19 is', np.mean(df['Overall']))
worst = df.loc[df['Overall'].idxmin()]
print('The worst players is',df['Overall'].min(), "overall. His name is", worst['Name'], 
      'he is from',worst['Nationality'],'and plays for',worst['Club'],'.')
```

```python
# DistPlot on Age

mage=df['Age'].mean()
meage=df['Age'].median()
sns.distplot(df['Age'])
plt.axvline(mage,color='r')
plt.axvline(meage,color='y')
plt.grid(True)

# This data is a positively skewed data
```

```python
df['Position'].value_counts()

# Most players position is ST
# Least players position is LF
```

```python
# Correlation

df.corr()
```

```python
# HeatMap

plt.figure(figsize=(10,10))
corr=df[['Potential','Name','Age','Overall','Strength','Acceleration']].corr()
sns.heatmap(corr,annot=True)

# Potential and age are negatively correlated
# Potential and Overall are highly correlated
# Strength and age are positively correlated
```

```python
# Highly correlated columns

plt.figure(figsize=(10,8))
corr=df[['Potential','Name','Age','Overall','Strength','Acceleration']].corr()
a=sns.heatmap(corr,annot=True,mask=(corr<0.6))
```

```python
# Negatively correlated columns

plt.figure(figsize=(10,8))
corr=df[['Potential','Name','Age','Overall','Strength','Acceleration']].corr()
a=sns.heatmap(corr,annot=True,cmap='Blues',mask=(corr>0))
```

```python
# Countplot on Nationality

plt.figure(figsize=(10,30))
sns.countplot(data=df,y='Nationality')
plt.grid(True)

# Most no of players nationality is england 
# Country with the least no of players is Fiji,Eritrea,Guatemala,Oman and Hong Kong       
```

```python
# PieChart on Preferred Foot

df['Preferred Foot'].value_counts(normalize=True).plot(figsize=(10,8),kind='pie',autopct='%.2f%%')
plt.show()

# 76.81% Players preferred foot is right
# 23.19% Players preferred foot is left
```

```python
# Subplot on Position and Age

plt.figure(1)
plt.subplot(2,2,1)
df['Position'].value_counts(normalize=True).plot(figsize=(10,8),kind='bar',color='red')
plt.title('Number of players position frequency diagram')
plt.xlabel('Position')
plt.ylabel('No. of players')

plt.subplot(2,2,2)
df['Age'].value_counts(normalize=True).plot(figsize=(12,10),kind='bar',color='green')
plt.title('Number of drive wheels frequency diagram')
plt.xlabel('Number of drive wheels')
plt.ylabel('drive wheels')

# Most preferred position is ST and least preferred position is LF
# Players with age 21 are more compared to others
```

```python
sns.scatterplot(df['Age'] , df['Overall'])
plt.title('Age vs Overall rating')
plt.grid(True)
plt.show()

# Players rated between 90-80 are mostly from 25 to 30 age bracket compared to other age group
```

```python
# Age Distribution

df=df[df['Nationality'].isin(['England', 'Germany', 'Spain'])]
sns.boxplot(x= df['Nationality'], y = df['Age'])
plt.show()
```
